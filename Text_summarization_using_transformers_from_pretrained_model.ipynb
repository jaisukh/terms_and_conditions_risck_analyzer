```python
import tensorflow as tf
import transformers
from transformers import T5Tokenizer,TFT5ForConditionalGeneration
```

```python
pretrained_model='t5-base'
tokenizer=T5Tokenizer.from_pretrained(pretrained_model)
model=TFT5ForConditionalGeneration.from_pretrained(pretrained_model)
```

```python
text="""As AI becomes more pervasive, we can expect it to take on mundane and tedious tasks that require a lot of human effort and time.
 One such activity is legal reviews of SaaS contracts between vendors and enterprise companies. Lawyers are quite expensive (ranging from 500$-1000$ an hour)
 and it would be great to have an AI agent conduct legal review of contracts and flag the risks to the buyer of SaaS within minutes. As part of the jazzee marketplace,
  when the buyer is about to purchase a product, he/she needs to sign off the Terms and Conditions. We need the AI conversational bot Janie to quickly scan the T&C and
  outline any risks back to the buyer so he/she can click OK and move on with the purchase or push it back to the vendor for changes."""
```

```python
tokenized_text=tokenizer(text,padding=True,truncation=True,return_tensors='tf',max_length=512)
generated_ids=model.generate(tokenized_text['input_ids'],max_length=1000, num_beams=4,early_stopping=True)
```

```python
summary=tokenizer.decode(generated_ids[0],skip_special_tokens=True)
```
